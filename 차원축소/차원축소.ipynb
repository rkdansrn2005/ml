```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
iris = load_iris()
iris.feature_names
columns=['sepal length','sepal width','petal length','petal width']
iris_df = pd.DataFrame(iris.data, columns = columns)
iris_df['target']= iris.target
iris_df.head(2)
```

```python
iris_df['target']==0
```

```python
markers=['^','s','o']
for i, marker in enumerate(markers):
    x_axis_data = iris_df[iris_df['target']==i]['sepal length']
    y_axis_data = iris_df[iris_df['target']==i]['sepal width']
    plt.scatter(x_axis_data, y_axis_data, marker=marker)
    
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
iris_scaled = StandardScaler().fit_transform(iris_df.iloc[:,:-1])
```

```python
iris_scaled.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(iris_scaled)
# 이걸 활용해서 나중의 테스트 데이터나 실제 데이터를 사용
iris_pca = pca.transform(iris_scaled)
print(iris_pca.shape)
```

```python
pca_columns=['pca_component_1','pca_component_2']
iris_df_pca = pd.DataFrame(iris_pca,columns=pca_columns)
iris_df_pca['target']= iris.target
iris_df_pca.head()
```

```python
markers=['^','s','o']
for i, marker in enumerate(markers):
    x_axis_data = iris_df_pca[iris_df_pca['target']==i]['pca_component_1']
    y_axis_data = iris_df_pca[iris_df_pca['target']==i]['pca_component_2']
    plt.scatter(x_axis_data, y_axis_data, marker=marker)
    
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
print(pca.explained_variance_ratio_)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
ref = RandomForestClassifier(random_state=156)
scores = cross_val_score(ref,iris.data,iris.target,scoring='accuracy',cv=3)
print(scores)
print(np.mean(scores))
```

```python
scores = cross_val_score(ref,iris_df_pca[['pca_component_1','pca_component_2']],iris.target,scoring='accuracy',cv=3)
print(scores)
print(np.mean(scores))
```

```python
df = pd.read_excel('credit_card.xls',sheet_name='Data',header=1)
```

```python
df.head(2)
```

```python
y_target = df.iloc[:,-1:]
X_features = df.drop(['ID','default payment next month'],axis=1)
```

```python
y_target.value_counts()
```

```python
X_features.info()
```

```python
import seaborn as sns
corr = X_features.corr()
plt.figure(figsize=(14,14))
sns.heatmap(corr,annot=True,fmt='.1g')
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#BILL_AMT1 ~ BILL_AMT6까지 6개의 속성명 생성
cols_bill = ['BILL_AMT'+str(i) for i in [1,2,3,4,5,6]]
print('대상 속성명:', cols_bill)
scaler = StandardScaler()
df_col_scaled = scaler.fit_transform(X_features[cols_bill])
pca = PCA(n_components=2)
pca.fit(df_col_scaled)
print(pca.explained_variance_ratio_)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
rcf = RandomForestClassifier(n_estimators=300,random_state=156)
scores= cross_val_score(rcf,X_features,y_target,scoring='accuracy',cv=3)
print(scores)
print(np.mean(scores))
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 원본 데이터셋에 먼저 StandardScaler적용
scaler = StandardScaler()
df_scaled = scaler.fit_transform(X_features)

# 6개의 Component를 가진 PCA 변환을 수행하고 cross_val_score( )로 분류 예측 수행. 
pca = PCA(n_components=6)
df_pca = pca.fit_transform(df_scaled)
scores_pca = cross_val_score(rcf, df_pca, y_target, scoring='accuracy', cv=3)

print('CV=3 인 경우의 PCA 변환된 개별 Fold세트별 정확도:',scores_pca)
print('PCA 변환 데이터 셋 평균 정확도:{0:.4f}'.format(np.mean(scores_pca)))
```
