```python
from scipy import stats

x = np.random.normal(0,1,size=30)

bandwidth=1.06*x.std()*x.size**(-1/5.)
print(bandwidth)
support = np.linspace(-4,4,200)
print(support)
```

```python
kernels=[]
for x_i in x:
    kernel= stats.norm(x_i,bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support,kernel,color='r')
print(kernels)    
sns.rugplot(x,color='.2',linewidth=3)
```

```python
from scipy.integrate import trapz
density = np.sum(kernels, axis=0)
density /= trapz(density, support)
plt.plot(support, density);
```

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift

X, y = make_blobs(n_samples=200, n_features=2, centers=3, 
                  cluster_std=0.8, random_state=0)
meansh = MeanShift(bandwidth=0.9)
custer_labels = meansh.fit_predict(X)
print(np.unique(custer_labels))
```

```python
meansh = MeanShift(bandwidth=1)
custer_labels = meansh.fit_predict(X)
print(np.unique(custer_labels))
```

```python
# estimate_bandwidth()      ,  bandwith  
from sklearn.cluster import estimate_bandwidth

bandwidth = estimate_bandwidth(X,quantile=0.25)
print('bandwidth :', round(bandwidth,3))
```

```python
meansh = MeanShift(bandwidth=1.689)
custer_labels = meansh.fit_predict(X)
print(np.unique(custer_labels))
```

```python
import pandas as pd

clusterDF = pd.DataFrame(data=X, columns=['ftr1', 'ftr2'])
clusterDF['target'] = y

# estimate_bandwidth()  bandwidth 
best_bandwidth = estimate_bandwidth(X, quantile=0.25)

meanshift= MeanShift(best_bandwidth)
cluster_labels = meanshift.fit_predict(X)
print('cluster labels :',np.unique(cluster_labels))  
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

clusterDF['meanshift_label']  = cluster_labels
centers = meanshift.cluster_centers_ #   
unique_labels = np.unique(cluster_labels)
markers=['o', 's', '^', 'x', '*']

for label in unique_labels:
    label_cluster = clusterDF[clusterDF['meanshift_label']==label]
    center_x_y = centers[label]
    
    #   marker scatter plot 
    plt.scatter(x=label_cluster['ftr1'], y=label_cluster['ftr2'], edgecolor='k', 
                marker=markers[label] )
    
    #   
    plt.scatter(x=center_x_y[0], y=center_x_y[1], s=200, color='white',
                edgecolor='k', alpha=0.9, marker=markers[label])
    plt.scatter(x=center_x_y[0], y=center_x_y[1], s=70, color='k', edgecolor='k', 
                marker='$%d$' % label)
    
plt.show()
```

```python
print(clusterDF.groupby('target')['meanshift_label'].value_counts())
```
